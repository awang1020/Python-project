# **2. Données**

Pour ce tutorial, nous allons utiliser Iris dataset. c’est un jeu de données qui contient 3 classes avec 50 enregistrements de chaque classe, et une classe désigne un type de la plante d’Iris.

Nous verrons dans un premier temps l’implémentation d’un modèle de machine learning classique puis nous allons voir comment l’autoML permet de nous donner un modèle et de tuner notre modèle.

On récupère les données grâce à la fonction **get_data()**.


```python
from pycaret.datasets import get_data
dataset = get_data('iris')
```

Ce jeu de données contient une description de la longueur et la largeur des sépales, la longueur et la largeur des pétales et le type d'Iris.

-----

Ensuite, pour réaliser la prédiction, nous allons déterminer une proportion de données pour faire l'entrainement et l'autre partie pour tester et réaliser la prédiction.

```python
data = dataset.sample(frac=0.9, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Données pour le modèle : ' + str(data.shape))
print('Données pour la prédiction: ' + str(data_unseen.shape))
```

On a donc pris 90 % de notre jeu pour l'entrainement et 10% pour tester notre modèle.

# **3. Set up l'environnement PyCaret**

Maintenant, nous allons setup notre environnement dans PyCaret.

La fonction **setup()** est l'initialisation de l'environnement et crée des pipelines de transformations et préparations de données pour la modélisation et le déploiement.

Il prend en paramètre un dataframe pandas et une colonne cible.

Lorsque setup() est exécuté, **le type des variables sera déduit par PyCaret**, il nous demandera de valider le type en appuyant sur la touche "Entrée". Si ce n'est pas le cas, vous pouvez utiliser les paramètres numeric_features, categorical_features, date_features dans la fonction de configuration pour les spécifier.

S'assurer que les types de données sont corrects est d'une importance fondamentale dans PyCaret car il effectue automatiquement quelques tâches de prétraitement qui sont impératives pour toute expérience d'apprentissage automatique.


```python
from pycaret.classification import *
exp_mclf101 = setup(data = data, target = 'species', session_id=1) 
```

*session_id* : Un nombre pseudo-aléatoire distribué comme seed dans toutes les fonctions pour une reproductibilité ultérieure. Ici, il est égal à 1.

# **4. Comparaison des modèles**

compare_models() est une fonction de PyCaret pour évaluer les performance de différents modèles, c'est le point de départ recommandé pour créer notre modèle.

La sortie imprime une grille de score qui affiche l'exactitude, le rappel, la précision, F1, Kappa et MCC par pli.

```python
best_model = compare_models(sort = 'Recall')
```

Cette grille est essentiel pour choisir notre modèle.
Ici, par défaut, les modèles sont trié par l'accuracy, on peut très bien trier par Recall en faisant

compare_models(sort = 'Recall')

# **5. Création d'un modèle**

Prenons un exemple en créant un modèle.

create_model() est une fonction de PyCaret et permet de créer le modèle. Il est souvent à la base de la plupart des fonctionnalités de PyCaret. Comme son nom l'indique, cette fonction forme et évalue un modèle à l'aide de la cross validation qui peut être définie avec le paramètre fold. La sortie imprime une grille de score qui affiche l'exactitude, le recall, la précision, F1, Kappa et MCC par pli.

---
Prenons l'exemple de la régression logistique, avec un fold de 5, c'est à dire une validation croisée de 5 fois :

```python
lr = create_model('lr', fold = 5)
```

On voit qu'on a une **moyenne** de 0.9573 et un **écart type **de 0.0615.


# **6.0 Tuner un modèle**

Maintenant, essayons d'avoir une meilleur précision en utilisant les hyperparamètres.

Lorsqu'un modèle est créé à l'aide de la fonction create_model(), il utilise les hyperparamètres par défaut pour entraîner le modèle. 

Afin d'ajuster les hyperparamètres, la fonction **tune_model()** est utilisée.

 La sortie imprime une grille de score qui affiche la précision, l'AUC, le recall, la précision, la F1, le Kappa et le MCC pour le meilleur modèle.

```python
tuned_lr = tune_model(lr)
```

Par défaut, il est réglé pour optimisé l'accuracy du modèle.

Les metrics (accuracy, recall...) seules ne sont pas les seuls critères à prendre en compte lors de la finalisation du meilleur modèle pour la production. D'autres facteurs à prendre en compte incluent le temps de formation, l'écart type des kfolds,...

# **7. Finaliser le modèle**

Enfin, finalisons le modèle

La finalisation du modèle est la dernière étape de l'expérience. 


La fonction **finalize_model()** adapte le modèle à l'ensemble de données complet, y compris sur l'échantillon test.


```python
final_lr = finalize_model(tuned_lr)
```

Affichons le modèle finale de machine learning multiclass pour ce jeu de données.

```python
print(final_lr)
```

# **8. Prédicition sur les données de test**

Appliquons ce modèle à nos données test qui contient 10 % de l'ensemble des données d'origine : nous allons utiliser la fonction **predict_model()**



```python
unseen_predictions = predict_model(final_lr, data=data_unseen)
unseen_predictions.head()
```

Les 2 dernières colonnes sont rajoutés : Label et Score.

Le label est la prédiction et le score est la probabilité de la prédiction.


# **9.0 Sauvegarde du modèle**

Nous avons maintenant terminé l'expérience en finalisant le modèle tuned_lr qui est maintenant stocké dans la variable final_lr. 

Nous avons également utilisé le modèle stocké dans final_lr pour prédire data_unseen. 

Ceci nous amène à la fin de notre expérience, mais une question reste à poser : que se passe-t-il lorsque vous avez de nouvelles données à prédire ? Devez-vous refaire toute l'expérience ? 

La réponse est non, la fonction intégrée save_model() de PyCaret vous permet d'enregistrer le modèle avec l'ensemble du pipeline de transformation pour une utilisation ultérieure.



```python
save_model(final_lr,'Final LR Model 19Nov2022')
```

# **10.0 Charger le modèle sauvegardé**

Pour charger un modèle enregistré à une date ultérieure dans le même environnement ou dans un environnement alternatif, nous utiliserions la fonction **load_model()** de PyCaret, puis appliquerions facilement le modèle enregistré sur de nouvelles données pour la prédiction.

```python
saved_final_ln = load_model('Final LR Model 11Nov2022')
```

Une fois le modèle chargé dans l'environnement, vous pouvez simplement l'utiliser pour prédire de nouvelles données à l'aide de la même fonction predict_model().

```python
new_prediction = predict_model(saved_final_ln, data=data_unseen)
new_prediction.head()
```

PyCaret est vraiment un bon outil d'autoML.

Nous avons vu qu'un workflow de machine learning dans PyCaret commence par setup(), suivi de la comparaison de tous les modèles à l'aide de compare_models() et du choix de modèle selon les critères de sélection (recall, accuracy...) pour ensuite, effectuer plusieurs techniques de modélisation telles que le réglage des hyperparamètres.

Tout cela est fait à partir d'un fonction facilement utilisable sous PyCaret.

Ce workflow vous mènera éventuellement au meilleur modèle à utiliser pour faire des prédictions sur des données nouvelles et inédites. 

# **11.0 La fonction automl**


Enfin, je vais vous présenter la fonction automl().

Cette fonction permet de sélectionner le meilleur modèle vu avec la fonction compare_models() basé sur le critère de sélection de votre choix. 

Par défaut, si aucun paramètre n'est renseigné, ce sera l'accuracy.

```python
best = automl()
best
```

Faisons un print pour afficher le meilleur modèle avec les hyperparamètres.

```python
print(best)
```

Faisons la prédiction sur les données test avec ce best_modèle : 

```python
predict_new = predict_model(best, data=data_unseen)
predict_new.head()
```

Le label est la prédiction et le score est la probabilité de la prédiction.

Nous voyons un score + élevé par rapport à notre modèle de régression logistique.
